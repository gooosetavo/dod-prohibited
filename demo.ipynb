```python
# Enclomiphene
# https://www.opss.org/dod-prohibited-dietary-supplement-ingredients
# https://www.reddit.com/r/navy/comments/1pza544/the_secret_list_thats_ending_military_careers/?share_id=O9uuE5Prlbqt1gKyshn6T&utm_content=2&utm_medium=ios_app&utm_name=iossmf&utm_source=share&utm_term=22
import requests
from bs4 import BeautifulSoup
import json


def fetch_drupal_settings(url):
    # Fetch the HTML document
    response = requests.get(url)
    response.raise_for_status()

    # Parse HTML
    soup = BeautifulSoup(response.text, "html.parser")

    # Find the script tag with the specific selector
    script_tag = soup.find(
        "script",
        {"type": "application/json", "data-drupal-selector": "drupal-settings-json"},
    )

    if not script_tag:
        raise ValueError("Drupal settings script tag not found")

    # Parse the JSON content
    settings = json.loads(script_tag.string)

    return settings


# Usage
url = "https://www.opss.org/dod-prohibited-dietary-supplement-ingredients"
settings = fetch_drupal_settings(url)


def get_nested(data, path, default=None):
    """Get nested value using dot notation: 'top.second.third'"""
    keys = path.split(".")
    value = data

    try:
        for key in keys:
            value = value[key]
        return value
    except (KeyError, TypeError):
        return default


prohibited_list = get_nested(settings, "dodProhibited")
# prohibited_list
```
